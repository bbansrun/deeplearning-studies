## Linear Regression

기존 데이터를 활용하여 학습을 시켜 regression model을 생성함
x,y값의 pair데이터

x -> regression model -> y


#### 가설 생성 (linear) hypothesis
많은 타입들이 linear 하게 그래프를 그림
'데이터 셋들이 linear한 선을 그릴 것이다.' 라는 가정

* Hypothesis
h(x) = Wx + b


#### which hypothesis is better

기존 데이터 셋들이
H(x) = Wx + b와 거리가 얼마나 다른가

-> Cost function
(X) H(x) - y
(O) (H(x) - y)^2 
제곱을 해서 음수가 아닌 동일한 위치에서 비교가 가능해짐

* 전체 데이터 셋의 그래프와의 거리 / 데이터 갯 수

minimize cost(W,b)
W와 b를 최소화 하는 경우의 수를 구하는 것이 목표

```python
import tensorflow as tf

# x and y data
x_train = [1,2,3]
y_train = [1,2,3]

#텐서플로 변수 선언
W = tf.Variable(tf.random_normal([1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')

#가설 정의
hypothesis = x_train*W + b

#reduce란 텐서가 주어졌을 때의 텐서들의 평균을 구해줌
cost = tf.reduce_mean(tf.square(hypothesis - y_train))

#minimize
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
#그래프 상의 노드의 이름
train = optimizer.minimize(cost)

#그래프를 만듬
sess = tf.Session()
#전체 변수를 그래프에 입력함
sess.run(tf.global_variables_initializer())

for step in range(2001):
    sess.run(train)
    if step % 20 == 0:
        print(step, sess.run(cost), sess.run(W), sess.run(b))
```

```python
X = tf.placeholder(tf.float32)
Y= tf.placeholder(tf.float32)

#feed_dict를 활용하여 값을 나중에 던져줌
for step in range(2001):
    cost_val, W_val, b_val, _= sess.run([cost,W,b,train], feed_dict={X:[1,2,3,4,5],Y:[2.1 , 3.1 , 4.1, 5.1, 6.1]})
    
    if step % 20 == 0:
        print(step, cost_val,W_val, b_val)
```
